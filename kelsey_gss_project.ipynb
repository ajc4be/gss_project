```python
import pandas as pd
import numpy as np

var_list = ['wrkstat', 'occ10', 'prestige', 'whatslf2', 'marital', 'divorce', 'martype',
            'sppres10', 'prestg10', 'spocc10', 'hapmar', 'age', 'educ', 'sex', 'income16', 'income'] # List of variables you want to save
output_filename = 'selected_gss_data.csv' # Name of the file you want to save the data to

modes = ['w','a'] # Has write mode and append mode
phase = 0 # Starts in write mode; after one iteration of loop, switches to append mode

for k in range(3): # for each chunk of the data
    url = 'https://github.com/DS3001/project_gss/raw/main/gss_chunk_' + str(1+k) + '.parquet' # Create url to the chunk to be processed
    print(url) # Check the url is correct
    df = pd.read_parquet(url) # Download this chunk of data
    print(df.head()) # Visually inspect the first few rows
    df.loc[:,var_list].to_csv(output_filename, # specifies target file to save the chunk to
                              mode=modes[phase], # control write versus append
                              header=var_list, # variable names
                              index=False) # no row index saved
    phase = 1 # Switch from write mode to append mode
```

**Cleaning prestg10**

prestg10 includes values from 0.0 to 100.0, based on the standardized score of how pretigious the respondent rated their occupation. Converted from object to float type.

```python
prestg10.unique()
```

```python
# Converting to numeric

prestg10 = pd.to_numeric(prestg10, errors="coerce")
print(prestg10.isna().sum()/len(prestg10))
```

prestg10 has ~7% missing values.

**Cleaning income**

```python
#type(income[0])
income = income.replace("income", np.nan) # replaced values euqal to "income" with NaN values
income.value_counts()
```

Making a copy of income with numerical values for data visualization and analysis purposes. Rounding each income category to its highest bound (e.g. 10,000-14,999 --> 14999)

```python
income_numeric = income.copy()
income_numeric = income_numeric.replace({"$25,000 or more": 25000,
                                         "$10,000 to $14,999": 12500,
                                         "$20,000 to $24,999": 22500,
                                         "$15,000 to $19,999": 17500,
                                         "$8,000 to $9,999": 9000,
                                         "$1,000 to $2,999": 2000,
                                         "$7,000 to $7,999": 7500,
                                         "$5,000 to $5,999": 5500,
                                         "$3,000 to $3,999": 3500,
                                         "$6,000 to $6,999": 6500,
                                         "$4,000 to $4,999": 4500,
                                         "under $1,000": 1000})

# There's a weird distribution of the values. Correcting it:
# Note: I did this by assigning each numeric value to its closest value of $0 to $25,000 in increments of 2500.
# For example, 4500 --> 5000.

income_numeric = income_numeric.replace({9000: 10000,
                                         6500: 7500,
                                         5500: 5000,
                                         3500: 2500,
                                         2000: 25000,
                                         1000: 0,
                                         4500: 5000})

income_numeric.value_counts()
```

**Cleaning wrkstat**

```python
# Replacing 'wrkstat' with NaN
wrkstat = wrkstat.replace('wrkstat', np.nan)

# Replacing 'with a job, but not at work because of temporary illness, vacation, strike' with 'temporary leave'
wrkstat = wrkstat.replace('with a job, but not at work because of temporary illness, vacation, strike', 'temporary leave')
# don't know if this is the best description - may want to change

wrkstat.value_counts()

missing_wrkstat_count = wrkstat.isna().sum()
percent_wrkstat = (missing_wrkstat_count / row_count) * 100
print(percent_wrkstat, "% of the wrkstat rows are NaN")
```

Only ~5% of wrkstat values missing.
